## symmetric binning

```python
nbins = 4

n_bin_edges = int(nbins / 2 - 1)
n_bin_edges
```

```python
bin_edges = [np.random.randint(1,20)/20 for i in range(n_bin_edges)]
```

```python
bin_edges = np.array(bin_edges)
bin_edges = np.append(bin_edges,[1])
bin_edges = np.append(bin_edges, bin_edges*-1)
bin_edges
```

```python
nbins = 10
n_bin_edges = int(nbins / 2 - 1)
ncomb = np.math.factorial(19) / (np.math.factorial(n_bin_edges) * np.math.factorial(19-n_bin_edges))
binnings = []
print(ncomb)

while len(binnings) < ncomb:
    bin_edges = np.array([np.random.randint(1,20)/20 for i in range(n_bin_edges)])
    bin_edges = np.append(bin_edges,[1])
    bin_edges = np.append(bin_edges, bin_edges*-1)

    if np.unique(bin_edges).shape == bin_edges.shape:
        binning = list(np.sort(np.append(bin_edges, [0])))
        if binning not in binnings:
            binnings.append(binning)
            print(binning)        
```

```python
binning = np.array([-1.0, -0.9, -0.65, -0.55, -0.25, 0.0, 0.25, 0.55, 0.65, 0.9, 1.0])
rebin = [ np.sum(bins[np.logical_and(x > binning[i], x < binning[i+1])]) for i in range(len(binning)-1) ]
rebin
```

```python
mask = (np.sqrt(rebin) / rebin) < 0.2
np.all(mask)
```

## non-symmetric binning

```python
nbins = 4

n_bin_edges = nbins - 2 - 1
n_bin_edges
```

```python
bin_edges_up = np.random.randint(1,20, n_bin_edges)/20
bin_edges_down = np.random.randint(-19,0)/20
```

```python
bin_edges = np.append(bin_edges_up, bin_edges_down)
bin_edges = np.append(bin_edges,[1, -1, 0])
np.sort(bin_edges)
```

```python
nbins = 5
n_bin_edges = nbins - 2 -1
ncomb_up = np.math.factorial(19) / (np.math.factorial(n_bin_edges) * np.math.factorial(19-n_bin_edges))
ncomb_down = np.math.factorial(19) / np.math.factorial(18)
print(ncomb_up, ncomb_down)
ncomb_up * ncomb_down
```
